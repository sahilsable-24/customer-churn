## Analyzing target variable

```python
clean_df["Churn"].value_counts()
```

```python
clean_df['Churn'].value_counts(normalize=True) * 100
```

```python
sns.countplot(x='Churn', data=clean_df)
plt.title("Churn distribution")
plt.show()
```

## Analyzing numerical columns

```python
numerical_cols = ["tenure", "MonthlyCharges", "TotalCharges"]

# distribution
clean_df[numerical_cols].hist(bins=30, figsize=(12,4))
plt.tight_layout()
plt.show()
```

```python
# numerical columns vs Churn

fig, axes = plt.subplots(1,3, figsize=(15,5))
for i, col in enumerate(numerical_cols):
    sns.barplot(x='Churn', y=col, data=clean_df, ax=axes[i])

plt.tight_layout()
plt.show()
```

### Customers who churn tend to have lower tenure and higher monthly charges.

## Analyzing categorical values


```python
clean_df.columns
```

```python
categorical_cols = ['gender', 'SeniorCitizen', 'Partner', 'Dependents',
                    'PhoneService', 'InternetService', 'Contract',
                    'PaymentMethod', 'TechSupport', 'OnlineSecurity']

fig, axes = plt.subplots(2, 5, figsize=(20, 10))
axes = axes.flatten()

for i, col in enumerate(categorical_cols):
    sns.countplot(x=col, hue='Churn', data=clean_df, ax=axes[i])
    axes[i].set_title(col)
    axes[i].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

### Customers on month-to-month contracts churn far more than those on 1 or 2 year contracts 

```python
### Heatmap
df_temp = clean_df.copy()
df_temp['Churn_binary'] = (df_temp['Churn'] == 'Yes').astype(int)

corr = df_temp[['tenure', 'MonthlyCharges', 'TotalCharges', 'Churn_binary']].corr()

sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

## Business Insights from EDA

1. ~27% of customers are churning — significant revenue loss risk
2. Month-to-month contract customers churn at much higher rates
3. Customers with shorter tenure are most at risk — early retention matters
4. Higher monthly charges correlate with more churn
5. Customers without OnlineSecurity or TechSupport churn more often
